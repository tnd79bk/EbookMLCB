```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np 

choose = 2

fig = plt.figure(figsize=(6, 6))
ax = fig.gca(projection='3d')
X, Y, Z = axes3d.get_test_data(0.05)

X = np.arange(-4, 4, 0.1)
Y = np.arange(-4, 4, 0.1)
X, Y = np.meshgrid(X, Y)
if choose == 1:
	Z = X**2 + Y**2
elif choose == 2: 
	Z = np.abs(X) + np.abs(Y) 
elif choose == 3: 
	Z = X**2 - Y**2
elif choose == 4:
	Z = 0.1*(X**2 + Y**2 - 10*np.sin(np.sqrt(X**2 + Y**2)))

# ax.plot_surface(X, Y, Z, rstride=8, cstride=8, cmap=cm.jet)
ax.plot_surface(X, Y, Z, cmap = 'gray')
offset = -2
cset = ax.contour(X, Y, Z, zdir='z', offset=offset, colors='black')
# cset = ax.contour(X, Y, Z, zdir='x', offset=0, cmap=cm.coolwarm)
# cset = ax.contour(X, Y, Z, zdir='y', offset=0, cmap=cm.coolwarm)

ax.set_xlabel('$x$', fontsize = 15)
ax.set_xlim(-4, 4)
ax.set_ylabel('$y$', fontsize =15)
ax.set_ylim(-4, 4)
ax.set_title('$f(x, y) = |x| + |y|$', fontsize = 14)
# if choose == 1:
# 	ax.set_zlabel('$f(x, y) = x^2 + y^2$', fontsize = 14)
# elif choose == 2:
# 	ax.set_zlabel('$f(x, y) = |x| + |y|$', fontsize = 14)
#     ax.set_title('$f(x, y) = |x| + |y|$', fontsize = 14)
# elif choose == 3:
# 	ax.set_zlabel('$f(x, y) = x^2 - y^2$', fontsize = 14)
# elif choose == 4:
# 	# ax.set_zlabel('$f(x, y) = \\frac{1}{10}(x^2 + 2y^2 - 2\sin(xy)$', fontsize = 20)
# 	ax.set_zlabel('$f(x, y)$', fontsize = 14)
# ax.set_title('$f(x, y) = \\frac{1}{10}(x^2 + y^2 - 10\sin(\sqrt{x^2 + y^2})$ (quasiconvex)', fontsize = 14)
ax.set_zlim(offset, 8)

from matplotlib.backends.backend_pdf import PdfPages
# filename = 'quasiconvex.pdf'
filename = 'norm1_surf.pdf'
with PdfPages(filename) as pdf:
    pdf.savefig(bbox_inches='tight')
plt.show()
```

```python
delta = 0.025
x = np.arange(-6.0, 5.0, delta)
y = np.arange(-20.0, 15.0, delta)
X, Y = np.meshgrid(x, y)
# Z = (X**2 + Y - 7)**2 + (X-Y + 1)**2

# Z = -X*np.log(0.8) - Y*np.log(0.2)
Z = np.maximum(np.abs(X) + 2*np.abs(Y), 2*X**2 + Y**2 - X*Y)

plt.figure(figsize=(6, 6))
CS = plt.contour(X, Y, Z, np.arange(0.5, 12, 1), colors = 'k')
plt.axis('equal')
plt.xlim(-3, 5)
plt.ylim(-3.5, 3.5)
plt.xticks([], [])
plt.yticks([], [])

plt.title('$f(x, y) = \max(2x^2 + y^2 -xy, |x| + 2|y|)$', fontsize = 20)
plt.axis('off')

filename = 'max_2d.pdf'
with PdfPages(filename) as pdf:
    pdf.savefig()  # saves the current figure into a pdf page
    
plt.show() 
plt.close()
```
